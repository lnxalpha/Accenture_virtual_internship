Faster method

```python
# Find the sum of each category
category_sums = remerged.groupby("Category")["Score"].sum()

# Find the top five highest categories
top_categories = category_sums.nlargest(5)

# Print the top five highest categories and their sums
print("Top 5 highest categories:")
for category, value in top_categories.iteritems():
    print(f"{category}: {value}")
    # Find the sum of individual category
    category_sum = remerged.loc[remerged["Category"].str.lower() == category, "Score"].sum()
    print(f"Sum of {category}: {category_sum}")
```
